```python
# bubble sort

class Bubble:
    def call(self,x):
        data = x.copy()
        for i in range(1,len(x)):
            for j in range(len(x)-i):
                if data[j] > data[j+1] :
                    data[j],data[j+1] = data[j+1],data[j]
                    
        return data
```

```python
# select sort

class Select:
    def call(self,x):
        data = x.copy()
        for i in range(len(x)):
            for j in range(i+1,len(x)):
                if data[i] > data[j]:
                    data[i],data[j] = data[j],data[i] 
        return data
```

```python
# insert sort

class Insert:
    def call(self,x):
        data = x.copy()
        for i in range(1,len(data)):
            index = i
            key = index[i]
            for j in range(i+1,-1,-1):
                if data[j] > key :
                    data[j+1] = data[j]
                    index = j
                    
                else : 
                    break
            data[index] = key
            
        return x
                    
```

```python
# shell sort

class Shell:
    def call(self, x):
        data = x.copy()
        mid = len(data)//2
        while mid > 0 :
            for i in range(mid,len(data)):
                index = i
                key = data[i]
                while index >= mid and data[index-mid] > key :
                    data[index] = data[index-mid]
                    index -= mid
                    
                data[index] = key
                
            mid//=2
                
        return data
```

```python
# radix sort

class Radix():
    def call(self,z):
        data = z.copy()
        isTrue = True
        radix = 1
        while isTrue:
            isTrue = False
            index = 0
            xlist = [[] for i in range(10)]
            
            for n in data:
                number = n//radix%10
                xlist[number].append(n)
            
                if (not isTrue) and  number > 0:
                    isTrue = True
            
            for dim2 in xlist:
                for dim1 in dim2:
                    data[index] = dim1
                    index+=1
            
            radix*=10
            
        return data
        
```

```python
# Stack

class Stack:
    
    def __init__(self, size=5):
        
        self.stack = []
        self.size = size
        self.top = 0
        self.check = True
        
    def push(self, data):
        self.check = True
        
        if self.top >= self.size :
            print("full stack")
            self.check = False
            return
        
        self.top+=1
        self.stack.append(data)
        print("push top = {} / {}".format(self.top, data))

    def pop(self):
        self.check = True
        if 0 ==  self.top:
            print("empty")
            self.check = False
            return
            
        print("pop top = {} / {}".format(self.top, self.stack[self.top-1]))
        self.stack.pop()
        self.top-=1
        
        
    def start(self,data):
        self.check = True
        index = 1
        while self.check and index < len(data):
            self.push(data[self.top])
            index += 1
            
        self.check = True    
        while self.check:
            self.pop()
```

```python
# Queue

class Queue:
    def __init__(self,size=5):
        self.Queue = []
        self.rear = 0
        self.front = 0
        self.size = size
        self.check = True
        
    def push(self,data):
        self.check = True
        if (self.front == self.rear-1) or (self.rear == 0 and self.front == (self.size-1)):
            print('full Queue')
            self.check = False
            return
        self.Queue.append(data)
        self.front = (self.front+1)%self.size
        print("push = {} / F{},R{} / {}".format(data,self.front,self.rear,self.Queue))
        
        
    def pop(self):
        self.check = True
        if self.front == self.rear :
            print('empty Queue')
            self.check = False
            return
        key = self.Queue[0]
        for i in range(len(self.Queue)-1):
            self.Queue[i] = self.Queue[i+1]
        self.Queue.pop()
        self.rear = (self.rear+1)%self.size
        print("push = {} / F{},R{} / {}".format(key,self.front,self.rear,self.Queue))
        
    def start(self,data):
        self.check = True
        index = 0
        while self.check and index < len(data)-1:
            self.push(data[index])
            index += 1
            
        self.check = True    
        while self.check:
            self.pop()
```

```python
# Binary Tree

class Binary:
    
    def __init__(self,data=None):
        self.data = data
        self.right = None
        self.left = None
    
    def insert(self,data):
        
        if self.data > data:
            
            if self.left is None :
                
                self.left = Binary(data)
                
            else :
                print("부모 왼쪽 노드 {} >> ".format(self.data))
                
                self.left.insert(data)
                
        elif self.data < data:
            
            if self.right is None :
                
                self.right = Binary(data)
                
            else :
                print("부모 오른쪽 노드 {} >> ".format(self.data))
                
                self.right.insert(data)
                
        else :
            print("이미 존재하는 값 입니다.")
    
    
    def inorder(self):
        
        
        if self.left : 
            self.left.preorder()
            
        print(self.data,end=' ')
        
        if self.right:
            self.right.preorder()
        
    def preorder(self):
        
        print(self.data,end=' ')
        if self.left : 
            self.left.preorder()
        if self.right:
            self.right.preorder()
            
    def postorder(self):
        if self.left : 
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.data,end=' ')
```
